```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
train_tokenized = tokenizer(
    train_tokens,
    is_split_into_words=True,
    padding=True,
    truncation=True
)

val_tokenized = tokenizer(
    val_tokens,
    is_split_into_words=True,
    padding=True,
    truncation=True
)

test_tokenized = tokenizer(
    test_tokens,
    is_split_into_words=True,
    padding=True,
    truncation=True
)

def align_labels(tokenized, labels):
    aligned = []

    for i, sent_labels in enumerate(labels):

        word_ids = tokenized.word_ids(batch_index=i)

        previous_word_idx = None
        label_ids = []

        for word_idx in word_ids:

            if word_idx is None:
                label_ids.append(-100)

            elif word_idx != previous_word_idx:
                label_ids.append(sent_labels[word_idx])

            else:
                label_ids.append(-100)

            previous_word_idx = word_idx

        aligned.append(label_ids)

    return aligned

train_tokenized["punct_labels"] = align_labels(train_tokenized, train_punct)
train_tokenized["cap_labels"] = align_labels(train_tokenized, train_cap)

val_tokenized["punct_labels"] = align_labels(val_tokenized, val_punct)
val_tokenized["cap_labels"] = align_labels(val_tokenized, val_cap)

test_tokenized["punct_labels"] = align_labels(test_tokenized, test_punct)
test_tokenized["cap_labels"] = align_labels(test_tokenized, test_cap)
```
