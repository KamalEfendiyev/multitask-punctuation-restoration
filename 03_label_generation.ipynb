```python
import re

def create_labels(sentence):

    tokens = sentence.split()

    punct_labels = []
    cap_labels = []
    clean_tokens = []

    for token in tokens:

        punct = "O"

        if token.endswith('.'):
            punct = '.'
        elif token.endswith(','):
            punct = ','
        elif token.endswith('?'):
            punct = '?'

        # убираем пунктуацию
        word = re.sub(r"[.,!?]", "", token)

        if len(word) == 0:
            continue

        # capitalization label
        if word.isupper():
            cap = "ALL"
        elif word[0].isupper():
            cap = "CAP"
        else:
            cap = "O"

        clean_tokens.append(word.lower())
        punct_labels.append(punct)
        cap_labels.append(cap)

    return clean_tokens, punct_labels, cap_labels
```
